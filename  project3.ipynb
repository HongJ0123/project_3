```python
import pandas as pd
import plotly.express as px
dog_license=pd.read_csv("https://data.cityofnewyork.us/resource/nu7n-tubp.csv")
canine_waste=pd.read_csv("https://data.cityofnewyork.us/resource/5npv-j6gn.csv")
```

```python
dog_license['licenseissueddate'] = dog_license['licenseissueddate'].astype(str)
dog_license['licenseexpireddate'] = dog_license['licenseexpireddate'].astype(str)

dog_license['licenseissueddate'] = dog_license['licenseissueddate'].str.replace('T', ' ', regex=False)
dog_license['licenseexpireddate'] = dog_license['licenseexpireddate'].str.replace('T', ' ', regex=False)

dog_license["licenseissueddate"] = pd.to_datetime(dog_license["licenseissueddate"], errors='coerce')
dog_license["licenseexpireddate"] = pd.to_datetime(dog_license["licenseexpireddate"], errors='coerce')
```

```python
canine_waste['installationdate'] = pd.to_datetime(canine_waste['installationdate'], errors='coerce')
canine_waste = canine_waste[canine_waste['installationdate'].notna()]
```

```python
dl_expire_2018 = dog_license[dog_license['licenseexpireddate'].dt.year == 2018]
```

```python
cw_2018 = canine_waste[canine_waste['installationdate'].dt.year == 2018]
```

```python
coerage_2018 = pd.merge(
    cw_2018[['zipcode', 'installationdate']],
    dl_expire_2018[['zipcode', 'licenseexpireddate']],
    on='zipcode',
    how='inner'
)
```

```python
installation_count_2018 = coerage_2018.groupby('zipcode').size().reset_index(name='installation_count_2018')
expire_license_count_2018 = dl_expire_2018.groupby('zipcode').size().reset_index(name='expired_license_count_2018')
```

```python
installation_per_licensed_dog = pd.merge(installation_count_2018, expire_license_count_2018, on='zipcode', how='outer')
```

```python
installation_per_licensed_dog = pd.merge(installation_per_licensed_dog, coerage_2018[['zipcode', 'installationdate']], on='zipcode', how='left')
installation_per_licensed_dog = pd.merge(installation_per_licensed_dog, dl_expire_2018[['zipcode', 'licenseexpireddate']], on='zipcode', how='left')
```

```python
installation_per_licensed_dog['installation_per_expired_license_ratio'] = (
    installation_per_licensed_dog['installation_count_2018'] /
    installation_per_licensed_dog['expired_license_count_2018']
)
```

```python
px.scatter(
    installation_per_licensed_dog,
    x='installationdate',
    y='installation_per_expired_license_ratio',
    color='zipcode',
    title='Monthly Installation per Expired License Ratio'
)
```
